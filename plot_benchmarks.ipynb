```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("delta_benchmark.csv")

# Read the benchmark data
# Each file with 128 cols holds 100MB of data
# With 10 files this is 1GB of data
for nappend in [0, 1, 3, 7, 9]:
    filtered_df = df[df['nappend'] == nappend]
    filtered_df.plot(x='ncols', y=['time_delta', 'time_pq'])
    plt.title(f'Read time with {nappend+1} files')
    plt.xlabel('ncols')
    plt.ylabel('read time (s)')
    plt.show()
```
